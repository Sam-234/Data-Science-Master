# Formating and Placeholder


```python
print("My name is samriddhi and age is 23")
```

```python
name = 'Samriddhi'
age = 23
print('My name is {} and age is {}'.format(name,age))
```

```python
#Placeholder
print('My name is {firstname} and age is {firstage}'.format(firstname = name,firstage=age))
```

```python
#with parameter
name = "Samriddhi"
age = 23
degree = 'Btech'
"My age is {fage} and I have completed: my degree of {fdegree} with the name of {fname}".format(fage =age,fdegree=degree,fname=name)
```

```python
price = int(input("Enter the product price:"))
if price > 1000:
    print('Get 20% off')
    print('The product price',price)
else:
    print('Get 30% off')
    print('The product price',price)
```
